**Box plots to show the salaries across different states**

```python
sns.set(rc={"figure.figsize":(30, 10)})
sns.boxplot(df_ds, x = 'Job Location', y = 'Avg Salary(K)')
```

X-axis: Represents different job locations. Each location will have its own box in the plot.
Y-axis: Shows the range of average salaries at each job location in the dataset.
Box: Each box in the boxplot represents the interquartile range (IQR) of the salaries for a particular location, with the line in the middle of the box depicting the median salary.
Whiskers: These extend from the box to show the range of the data, specifically to 1.5 times the IQR beyond the quartiles. Points beyond the whiskers indicate outliers or unusual salary values outside the typical range.
Color and Style: The default style of Seaborn ensures a clear and visually appealing representation.

```python
plt.pie(df_ds.Degree.value_counts(), labels = df_ds.Degree.value_counts().index
        ,autopct='%.0f%%')
```

The code uses Matplotlib to create a pie chart visualizing the distribution of degrees within the df_ds dataset. df_ds.Degree.value_counts() calculates the frequency of each unique value in the 'Degree' column, essentially counting how many entries exist for each type of degree. This frequency data is used to determine the size of each slice of the pie chart. The labels parameter assigns the names of the degrees to the respective slices. The autopct='%.0f%%' formats the percentage shown on each slice to a whole number, enhancing readability by displaying how much each degree type contributes to the total, expressed as a percentage.

## **Insights and Trends**

Indeed, the field of data science requires a diverse range of technical skills to excel in the profession

Regarding education levels, the requirements can vary depending on the industry and the specific job role. While some data science positions may require a bachelor's degree in fields like information technology, business services, others indutries like biotechnology and pharma may prefer advanced degrees such as a Ph.D. However, it's worth noting that practical experience and demonstrated skills often carry significant weight in the field of data science.

When it comes to salaries, they indeed vary significantly based on the location of the job and the type of company. Data scientists working in states like California, Massachusetts, Illinois tend to have higher salaries due to the higher cost of living and increased demand for data science talent whereas data scientists living in states such as Idaho, Nebraska, and Alabama demand lower salaries. Similarly, data scientists working for large corporations or well-established tech companies often command higher salaries compared to those working for smaller or non-profit organizations. Additionally, factors such as years of experience, level of expertise can also impact salary levels.
